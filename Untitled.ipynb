Note: This project is used for practice, i.e. showing the basic techniques of the data science project.

For starters, we’ll import some of the required python libraries, import the data, and take a little look at the data to get a first impression of them.

```python
data.drop("tax", axis=1, inplace=True)
```

Machine learning models require that the input variables be numerical. This means that if the data contains categorical data, we need to code them into numbers so that we can adjust and evaluate the model. We will use OHE (One-Hot Encoding).

```python
data_encoded = pd.get_dummies(data)
data_encoded
```

Let's scale the data.

```python
standard_scaler = StandardScaler()
data_encoded_scaled = standard_scaler.fit_transform(data_encoded)
data_encoded_scaled = pd.DataFrame(data_encoded_scaled, columns = data_encoded.columns)
data_encoded_scaled
```

We now arrange the data so that we take 80% of the data for training and 20% for testing.

```python
X_train, X_test, y_train, y_test = train_test_split(data_encoded_scaled.drop("price", axis=1), data_encoded_scaled[["price"]], test_size = 0.2, random_state = 0)
```

Let’s now explore which model is best suited for our case.

```python
list_of_models = [LinearRegression(), Ridge(), Lasso(), RandomForestRegressor(), GradientBoostingRegressor(), KNeighborsRegressor(), SVR()]

models_df = pd.DataFrame(columns = ["Model", "Score"])
models_df = models_df.append({"Model": "LINEAR", "Score": "/"}, ignore_index=True)

for model in list_of_models:
    regressor = model
    regressor.fit(X_train, y_train.values.ravel())
    score = regressor.score(X_test, y_test)
    models_df = models_df.append({"Model": model, "Score": score*100}, ignore_index=True)
    
models_df
```

We see that the RandomForestRegressor() model has a very high prediction accuracy (i.e. r^2 score)

Let's transform the data, and once again check how the selected models work.

```python
polynomial_features = PolynomialFeatures()
X_train_transformed = polynomial_features.fit_transform(X_train)
X_test_transformed = polynomial_features.transform(X_test)

models_df = models_df.append({"Model": "POLYNOMIAL", "Score": "/"}, ignore_index=True)

for model in list_of_models:
    regressor = model
    regressor.fit(X_train_transformed, y_train.values.ravel())
    score = regressor.score(X_test_transformed, y_test)
    models_df = models_df.append({"Model": model, "Score": score*100}, ignore_index=True)
    
models_df
```

After polynomial data transformation, we obtained the best score: r^2 = 95.969919%, and the model that gave this score is RandomForestRegressor().
